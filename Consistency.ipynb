```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn
```

```python
data_dir1 = "./test/vor"
data_dir2 = "./test/rbf"
data_dir3 = "./test/ags"
trials1 = ['log_vor.txt','log_vor2.txt','log_vor3.txt','log_vor4.txt']
trials2 = ['log_rbf.txt','log_rbf2.txt','log_rbf3.txt','log_rbf4.txt']
trials3 = ['log_ags.txt','log_ags2.txt','log_ags3.txt','log_ags4.txt']
title=['Gap on the right','Gap on the slight right','Gap on the left','Gap straight ahead']
for i,[f1,f2,f3] in enumerate(zip(trials1,trials2,trials3)):
    vor = pd.read_table(os.path.join(data_dir1,f1),header=None,names=["Voronoi + AGS"])[:70]
    rbf = pd.read_table(os.path.join(data_dir2,f2),header=None,names=["RBF + AGS"])[:70]
    ags = pd.read_table(os.path.join(data_dir3,f3),header=None,names=["AGS Only"])[:70]
    df = pd.concat([vor, rbf, ags], axis=1, )
    plt.figure()
    df.plot()
    plt.title(title[i])
    plt.xlabel("Time Steps")
    plt.ylabel("Turning Command")
    plt.show()
```
